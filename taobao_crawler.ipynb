```python
import requests as r
url = 'https://world.taobao.com/search/json.htm?navigator=all&_ksTS=1509682984796_28&spm=a213z.1224559.20141209.1&_input_charset=utf-8&json=on&q=%E7%8E%A9%E5%85%B7&cna=tJ5GEqAay3ECAW%2F79Z1gOc05&callback=__jsonp_cb&abtest=_AB-LR517-LR854-LR895-PR517-PR854-PR895&nid=&type=&uniqpid='
resp = r.get(url)
print(resp.text)
```

```python
import json
import re
m = re.search('if\(window.__jsonp_cb\)\{__jsonp_cb\((.*?)\)\}',resp.text)
print(m.group(1))
```

```python
jd=json.loads(m.group(1))
jd
```

```python
jd['itemList']
```

```python
with open('a.json','w') as f:
    f.write(json.dumps(jd))
```

```python
for item in jd['itemList']:
    print('商品名:' + item['title']+
          '\n'+'價格:'+item['price']+
          '\n'+item['nick'])
```
