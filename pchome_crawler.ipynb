```python
import requests as r
url = "http://ecshweb.pchome.com.tw/search/v3.3/?q=ssd"
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
resp = r.get(url,headers=headers)
resp.text
```

```python
#判斷應該是用Javascript
#到XHR|JS找到對的URL
url = 'http://ecshweb.pchome.com.tw/search/v3.3/all/results?q=ssd&page=1&sort=rnk/dc'
resp = r.get(url,headers=headers)
resp.text
```

```python
import json
jd=json.loads(resp.text)
jd
```

```python
for item in jd['prods']:
    print('商品名:' + item['name']+
          '\n'+'價格:'+str(item['price']))
```

```python
import pandas as pd
df = pd.DataFrame(jd['prods'])
df
```
