```python
import requests as r
from bs4 import BeautifulSoup as bs
import re
URL = 'http://www.coolpc.com.tw/evaluate.php'

resp =  r.get(URL)

s = bs(resp.text,'html5lib')
```

```python
#找尋class 名稱是'select',attribute是一個dict
#找到LABEL內HDD名子
hdds_optgroup = (s.find(name='select',attrs={'name':'n7'})
                 .findAll(lambda elem: ('HDD' in elem.get('label',' ')) and ('3.5' in elem.get('label',' ')))) 
                #LABEL後面這空格elem.get('label',' ') 是因為找不到label後要給他一個default值，沒有的話會噴錯

hdds_optgroup
```

```python
hdds = []
for optgroup in hdds_optgroup:
    hdd_text = [x.text for x in optgroup.findAll('option')]
    hdds.extend(hdd_text)

hdds
```

```python
hdd = 'Toshiba 1TB (DT01ACA100)/32M/7200轉/三年保, $1599 ◆ ★ ↓酷幣19↓\n'
```

```python
name = hdd.split('/')[0]
name 
```

```python
capacity = re.findall('\d{1,3}G|\d{1,3}TB',hdd)[0]
capacity
```

```python
cache = re.findall('\d{1,3}M|\d{1,3}MB',hdd)[0]
cache
```

```python
speed = re.findall('\d+轉',hdd)[0]
speed
```

```python
warranty = re.findall('.年',hdd)[0]
warranty
```

```python
price = re.findall(('\$(\d+)'),hdd)[0]
price
```

```python
hdd_list = []
for hdd in hdds:
    hdd_dict = {}
    hdd_dict['name'] = hdd.split('/')[0]
    
    capacity = re.findall('\d{1,3}G|\d{1,3}TB',hdd)[0]
    _cap = int(re.findall('\d+',capacity)[0])
    if 'TB' in capacity:
        hdd_dict['capacity'] = _cap * 1000
    else:
        hdd_dict['capacity'] = _cap
        
    hdd_dict['cache']=int(re.findall('\d+',re.findall('\d{1,3}M|\d{1,3}MB',hdd)[0])[0])
    hdd_dict['speed']=int(re.findall('(\d+)轉',hdd)[0])
    hdd_dict['warranty'] = re.findall('.年',hdd)[0]
    hdd_dict['price'] =int(re.findall(('\$(\d+)'),hdd)[0])
    
    hdd_list.append(hdd_dict)
```

```python
hdd_list
```

```python
import csv

with open('hdd,csv','w') as csvfile:
    writer = csv.DictWriter(csvfile,fieldnames=['name', 'capacity', 'cache', 'speed', 'warranty', 'price'])
    writer.writeheader()
    writer.writerows(hdd_list)
```

```python
import pandas as pd
df = pd.DataFrame(hdd_list)
df
```
